# Web Scraping Project – AmbitionBox


##  Step 1: Import Libraries


```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

## Step 2: Fetch Webpage

```python
headers= {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
webpage=requests.get("https://www.ambitionbox.com/list-of-companies?page=1",headers=headers).text
```

## Step 3: Parse HTML

```python
soup = BeautifulSoup(webpage,'lxml')
```

## Step 4: Extract Data

```python
company = soup.find_all('div', class_='companyCardWrapper')
name=[]
rating=[]
review=[]
salaries=[]
interviews=[]
jobs=[]
benefits=[]


for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('div', class_="rating_text").text.strip())
    review.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[0].text.strip())
    salaries.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[1].text.strip())
    interviews.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[2].text.strip())
    jobs.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[3].text.strip())
    benefits.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[4].text.strip())
    

d={'name':name, 'rating':rating,'review':review,'salaries':salaries,'interviews':interviews,'jobs':jobs,'benefits':benefits}    

df=pd.DataFrame(d)
df


    



```
